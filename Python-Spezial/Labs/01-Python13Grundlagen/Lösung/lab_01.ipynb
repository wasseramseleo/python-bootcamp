# Lösung Lab 01: Bank Transactions Analysis

### Basis Aufgabe

```python
# Gegebene Daten
transactions = [
    {"id": 101, "type": "deposit", "amount": 100.00, "currency": "EUR"},
    {"id": 102, "type": "withdrawal", "amount": 50.00, "currency": "EUR"},
    {"id": 103, "type": "deposit", "amount": 200.00, "currency": "USD"},
    {"id": 104, "type": "payment", "amount": 25.50, "currency": "EUR"},
    {"id": 105, "type": "withdrawal", "amount": 1000.00, "currency": "USD"},
]
usd_to_eur_rate = 0.90

# 1. Initialisierung
balance = 0.0
processed_amounts = []

# 2. Verarbeitungsschleife
for tx in transactions:
    current_amount = tx["amount"]
    
    # 3. Währungsumrechnung (USD -> EUR)
    if tx["currency"] == "USD":
        current_amount = current_amount * usd_to_eur_rate
    
    # 4. Logik (Deposit vs Withdrawal)
    if tx["type"] == "deposit":
        balance += current_amount
    elif tx["type"] in ["withdrawal", "payment"]:
        balance -= current_amount
        
    # 5. Speicherung für Teil 2
    processed_amounts.append(current_amount)

# 6. Output
print(f"Final Balance: {balance} EUR")
print(f"Processed amounts: {processed_amounts}")
```

### Bonus Herausforderung

```python
# 1. List Comprehension
# Konvertierung zu Integer (Ganzzahlen), analog zu "int_weights" im Theorie-Code
eur_only_int = [int(amount) for amount in processed_amounts]

print(f"Integer Amounts: {eur_only_int}")
```

```python
# 2. Generator Expression
# Lazy Evaluation: Berechnung passiert erst beim Zugriff (iterieren)
# Analog zu "energy_gen" im Theorie-Code
fee_gen = (amount * 0.01 for amount in processed_amounts)

print("Calculated Fees:")
for fee in fee_gen:
    print(f"{fee:.2f} EUR")
```

```python
# 3. Type Hinting
# Verwendung von Standard-Typen wie im Theorie-Block gezeigt

def check_limit(amount: float) -> bool:
    """Prüft ob eine Transaktion das Limit überschreitet."""
    return amount > 500.0

# Test der Funktion
print(f"Is 1000.0 over limit? {check_limit(1000.0)}")
print(f"Is 50.0 over limit? {check_limit(50.0)}")
```
